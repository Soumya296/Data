### Importing Library

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
%matplotlib inline

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

### Reading the Data

```python
data = pd.read_csv("Churn_Modelling.csv")
data
```

### Dropping unnecessary features

```python
print("As the RowNumber, CustomerID, Surname are not important features in predicting the churing, we are dropping it")
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
data
```

### Segregating Dependent and Independent Variables

```python
x = data[data.columns.drop('Exited')]
x
```

```python
dependent_col = ['Exited']
y = torch.tensor(data[dependent_col].to_numpy()).to(device)
y.shape
```

### One hot encoding of Categorical Features

### Geography

```python
country_one_hot = pd.get_dummies(data.Geography,prefix='Geography')
```

```python
x = x[x.columns.drop('Geography')]
x
```

```python
x = pd.concat([x,country_one_hot],axis=1)
x
```

### Gender

```python
gender_one_hot = pd.get_dummies(data.Gender,prefix='Gender')
gender_one_hot
```

```python
x = x[x.columns.drop('Gender')]
x
```

```python
x = pd.concat([x,gender_one_hot],axis=1)
x
```

### Scaling of Features

```python
scaling_col = [x.columns.drop(['HasCrCard','IsActiveMember','Geography_France', 'Geography_Germany','Geography_Spain','Gender_Female','Gender_Male'])]
scaling_col
```

```python
scale_x = x[x.columns.drop(['HasCrCard','IsActiveMember','Geography_France', 'Geography_Germany','Geography_Spain','Gender_Female','Gender_Male'])]
scale_x
```

```python
scaler = preprocessing.StandardScaler()
x_scaled = scaler.fit_transform(scale_x)
x_scaled
```

```python
x_scaled = pd.DataFrame(x_scaled, columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts','EstimatedSalary'])
x_scaled
```

```python
x[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts','EstimatedSalary']] = x_scaled
x
```

### Converting to Tensor and loading on device

```python
x = torch.tensor(x.to_numpy(),dtype=torch.float32) ##converting to the tensor
```

```python
x.to(device)
print(x.shape) #number of input nodes
```

### Splitting

```python
train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.3)
train_x = train_x.to(device)
test_x = test_x.to(device)
train_y = train_y.to(device)
test_y = test_y.to(device)
```

```python
print("Size of train_x : {}".format(train_x.shape))
print("Size of train_y : {}".format(train_y.shape))
print("Size of test_x : {}".format(test_x.shape))
print("Size of test_y : {}".format(test_y.shape))
```

### Model Definition

```python
class Net(nn.Module):
    
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(13, 32) 
        self.fc2 = nn.Linear(32, 128)
#         self.fc3 = nn.Linear(128, 512)
#         self.fc4 = nn.Linear(512,128)
        self.fc5 = nn.Linear(128, 64)
        self.fc6 = nn.Linear(64, 16)
        self.fc7 = nn.Linear(16, 1)
        self.dropout = nn.Dropout(p=0.2)
        self.batchnorm1 = nn.BatchNorm1d(32)
        self.batchnorm2 = nn.BatchNorm1d(128)
        self.batchnorm5 = nn.BatchNorm1d(64)
        self.batchnorm6 = nn.BatchNorm1d(16)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.batchnorm1(x)
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.batchnorm2(x)
        x = self.dropout(x)
        x = F.relu(self.fc5(x))
        x = self.batchnorm5(x)
        x = F.relu(self.fc6(x))
        x = self.batchnorm6(x)
        x = torch.sigmoid(self.fc7(x))
        return x


net = Net().to(device)
```

```python
y_pred = net(train_x).to(device)
y_pred.shape
```

### Loss Function

```python
loss_fn = nn.BCEWithLogitsLoss()
loss = loss_fn(y_pred,train_y.float())
loss
```

### Training Neural Network

### Adam

```python
epochs = 1200
optimizer = torch.optim.Adam(net.parameters())
Losses = []
Accuracy = []
```

```python
batch_size = 32
batch_id = torch.round(torch.tensor(train_x.shape[0]/batch_size))
print("Batch Size : {} | Total Batches : {}".format(batch_size,batch_id))

# batch_idx = range(batch_id.int())[-1]
# x_batch, y_batch = train_x[batch_idx*batch_size:,:],train_y[batch_idx*batch_size:,:]
# pred_y = net(x_batch.float()).to(device)
# loss = loss_fn(pred_y,y_batch.float())
# epoch_loss = 0
# epoch_loss += loss
# epoch_loss
# y_batch.shape
```

```python
for epoch in range(epochs+1):
    epoch_loss = 0
    Acc = 0
    for batch_idx in range(batch_id.int()):
        if batch_idx != batch_id-1:
            x_batch = train_x[batch_idx*batch_size:(batch_idx+1)*batch_size,:].to(device)
            y_batch = train_y[batch_idx*batch_size:(batch_idx+1)*batch_size,:].to(device)
        else:
            x_batch, y_batch = train_x[batch_idx*batch_size:,:].to(device),train_y[batch_idx*batch_size:,:].to(device)
        
        pred_y = net(x_batch.float()).to(device)
        loss = loss_fn(pred_y,y_batch.float()).to(device)
        epoch_loss += loss
        pred_y_round = torch.round(pred_y)
        Acc += (pred_y_round == y_batch).sum().float()/y_batch.shape[0]
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    Losses.append(epoch_loss)
    Accuracy.append(Acc/batch_id)
    if epoch%10 == 0:
        print("Epoch : {} \t| loss : {} \t| accuracy : {}".format(epoch,epoch_loss/batch_id,Acc/batch_id))
```

### Test Data

```python
net.eval()
with torch.no_grad():
    pred_y_test = torch.round(net(test_x.float()))
    loss = loss_fn(pred_y_test,test_y.float())
    Acc = (pred_y_test == test_y).sum().float()/test_y.shape[0]
    print("On Testing Data ==> \t loss : {} \t| accuracy : {}".format(loss,Acc))
```

```python
plt.plot(range(1,epochs+1),[Loss.cpu().detach().numpy() for Loss in Losses])
```

```python
plt.plot(range(1,epochs+1),[Acc.cpu().detach().numpy() for Acc in Accuracy],c='orange')
```

### Confusion Matrix

```python
confusion_matrix(pred_y_test.cpu(),test_y.float().cpu())
```

```python
tn, fp, fn, tp = confusion_matrix(pred_y_test.cpu(),test_y.float().cpu()).ravel()
print('True Negatives : {}'.format(tn))
print('False Negatives : {}'.format(fn))
print('True Positives : {}'.format(tp))
print('False Positives : {}'.format(fp))
```
